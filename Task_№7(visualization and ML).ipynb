# 2.Створіть дві гістограми по полю fico (перша для значень not.fully.paid=0, друга = 1). Гістограми накладіть одна на одну.

```python
not_fully_paid = df[df['not.fully.paid'] == 1]['fico']
fully_paid = df[df['not.fully.paid'] == 0]['fico']

kwargs = dict(histtype='stepfilled', alpha=0.5, bins=15)

sns.set_style("darkgrid")
plt.hist(fully_paid, label='fully paid', **kwargs)
plt.hist(not_fully_paid, label='not fully paid', **kwargs)

plt.xlabel('Кредитний рейтинг позичальника FICO')
plt.ylabel('Кількість клієнтів')
plt.title('Кредитний рейтинг позичальника FICO в залежності від загальної виплати')

plt.legend()
plt.show()

```

По даному графіку одразу видно, що кількість клієнтів які повністю виплатили кредит набагато більша,
розподіл по кредитному рейтингу майже однаковий:
кількість позичальників з рейтингом до 600 дуже мала або відсутня, 680-700 пік і майже плавнне спадання до рейтингу 800 і більше...

# 3.Відобразіть тенденцію між оцінкою fico та int.rate. Використайте jointplot

# 5.purpose є категорієї, тому варто перетворити за допомогою pd.get_dummies.

```python
data = pd.get_dummies(df, columns=["purpose"])
#розбиття колонки "purpose" на колонки з значеннями "0" та "1" для перетворення даних у цифрові значення.
```

```python
data.info()
#додалися колонки 13-19 включно замість "purpose"
```

# 6.Розбийте датасет на тестві та трейнові дані

```python
from sklearn.model_selection import train_test_split
X = data.drop('not.fully.paid', axis=1)
y = data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2023)
```

розбиття на тестові та трейнові значення у співідношенні: 25% та 75% відновідно.

# 7.Використайте для задачі класифікації такі моделі: дерево рішень, логістична регресія, random forest, XGBoost.Для кожної з моделей виведіть матрицю, основні метрики

```python
#Дерево рішень
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
tree = DecisionTreeClassifier(max_depth=27, random_state=2023)
tree.fit(X_train, y_train)

y_pred = tree.predict(X_test)
print(classification_report(y_test, y_pred))
```

Бачимо слабкі результати для розпізнавання позитивних випадків

```python
pd.DataFrame(confusion_matrix(y_test, y_pred),
             index=['True_0', 'True_1'],
             columns=['Pred_0', 'Pred_1'])
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
recall_scores = []
for depth in range(2, 50):
  tree = DecisionTreeClassifier(max_depth=depth, random_state=2023)
  tree.fit(X_train, y_train)
  y_pred = tree.predict(X_test)
  recall_scores.append(recall_score(y_test, y_pred))

plt.plot(range(2, 50), recall_scores)
plt.xlabel('Depth')
plt.ylabel('Recall score')
plt.show()
```

Побудував графік, на якому побачимо при якій кулькості гілок(глибина дерева) в моделі буде найвище значення recall.Щоб модель була здатна виявити всі доступні позитивні приклади, слід орієнтуватися на максимізацію Recall.

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter=1000, random_state=2023, class_weight={0: 1, 1: 5})
log_reg.fit(X_train, y_train)

y_pred = log_reg.predict(X_test)
print(classification_report(y_test, y_pred))
```

модель здатна виявити більше доступних позитивних випадків

```python
pd.DataFrame(confusion_matrix(y_test, y_pred),
             index=['True_0', 'True_1'],
             columns=['Pred_0', 'Pred_1'])
```

```python
#RandomForest
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=7, random_state=2023)
forest.fit(X_train, y_train)

y_pred = forest.predict(X_test)
print(classification_report(y_test, y_pred))
```

Дуже малий recall для позитивних випадків

```python
pd.DataFrame(confusion_matrix(y_test, y_pred),
             index=['True_0', 'True_1'],
             columns=['Pred_0', 'Pred_1'])
```

```python
recall_scores = []
for n_est in [10, 20, 50, 100]:
  forest = RandomForestClassifier(n_estimators=n_est, random_state=2023, n_jobs=-1)
  forest.fit(X_train, y_train)
  y_pred = forest.predict(X_test)
  recall_scores.append(recall_score(y_test, y_pred))

plt.plot([10, 20, 50, 100], recall_scores)
plt.xlabel('Number of trees')
plt.ylabel('Recall  score')
plt.show()
```

```python
#XGBoost
import xgboost as xgb
from xgboost import XGBClassifier
model = xgb.XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_test, y_pred),
             index=['True_0', 'True_1'],
             columns=['Pred_0', 'Pred_1'])
```

Дуже малий recall для позитивних випадків

# 8.Використайте ансамбель voting для всіх моделей

```python
from sklearn.ensemble import VotingClassifier
import xgboost as xgb
from xgboost import XGBClassifier
model1 = DecisionTreeClassifier(max_depth=27, random_state=2023)
model2 = LogisticRegression(max_iter=1000, random_state=2023, class_weight={0: 1, 1: 5})
model3 = XGBClassifier()
model4 = RandomForestClassifier(n_estimators=7, random_state=2023)
voting_model = VotingClassifier(estimators=[('dt', model1), ('lr', model2), ('xgb', model3), ('Rf', model4)])
voting_model.fit(X_train, y_train)
y_pred = voting_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

# 3.Яка середня конценрація  NO2  для кожного дня тижня і міста (виведіть таблицю)?

```python
ds['day_of_week'] = ds['date'].dt.day_name()
ds.groupby(["city", "day_of_week"])["value"].mean()
```

```python
ds.groupby("utc")["value"].mean().sort_values().plot(kind='bar')
plt.xlabel('Година')
plt.ylabel('Середнє значення NO2')
plt.title('Середнє значення NO2 для кожної години')
plt.show()
#не видно ніякої залежності концентрації NO2 від години
```
